# JSON to RTTM
Let's do some `json` to `RTTM` :o
Notes:
- `Channel ID` always `1`
- `Orthography Field` and onwards is `<NA>`


Assumes that the JSON is always in the format given by the task.

````python
import json

transcript_file = 'eleven/transcript.json'

with open(transcript_file, 'r') as f:
    data = json.load(f)
print(data)
````

Format is
```
    "ep-11": [
        {
            ...
            "speaker": "...",
            ...
            ...
        }
    ]
```

````python
with open('notebooks/transcript.rttm', 'w') as f_o:
    for segment in data['ep-11']:
        start = segment['utterance_start']
        duration = segment['duration']
        speaker = segment['speaker']
        speech = segment['utterance']
        line = f"SPEAKER 11 1 {start:.3f} {duration:.3f} <NA> <NA> {speaker.replace(" ", "_")} <NA> <NA>\n"
        f_o.write(line)
````
